```python
import plotly.express as px
import covid
import pandas as pd
from scipy import signal
import numpy as np
```

```python
df = covid.get_data()
dfs = []

for county in covid._io.BAYAREA_COUNTIES:
    filtercut = (df.county == county)
    df_temp = df[filtercut].copy()

    cases_per_day = np.concatenate(([0], np.diff(df_temp.cases)))
    filtered_cases = signal.savgol_filter(cases_per_day, 15, 3)
    deaths_per_day = np.concatenate(([0], np.diff(df_temp.deaths)))
    filtered_deaths = signal.savgol_filter(deaths_per_day, 15, 3)

    df_temp['cases_per_day'] = cases_per_day
    df_temp['filtered_cases'] = filtered_cases
    df_temp['deaths_per_day'] = deaths_per_day
    df_temp['filtered_deaths'] = filtered_deaths

    dfs.append(df_temp)
    
df_plot = pd.concat(dfs)
```

```python
fig = px.line(
    df_plot,
    x='date',
    y="cases",
    color='county',
    color_discrete_sequence=px.colors.qualitative.Set1,
)
fig.write_html('total_cases_by_county.html')
fig.show()
```

```python
fig = px.line(
    df_plot,
    x='date',
    y="filtered_cases",
    color='county',
    color_discrete_sequence=px.colors.qualitative.Set1,
)
fig.write_html('daily_cases_by_county.html')
fig.show()
```
